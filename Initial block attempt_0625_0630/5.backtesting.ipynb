# Backtesting

```python
import pandas as pd
import numpy as np
```

```python
back_df = pd.read_csv('model_out.csv')
back_df['date_time'] = pd.to_datetime(back_df['date_time'])
back_df.set_index('date_time', inplace=True)
back_df
```

```python
def strategy(pre_rtn):
    a = back_df['return_1h'].quantile(0.75)
    b = back_df['return_1h'].quantile(0.25)
    if pre_rtn > a:
        return pre_rtn
    elif pre_rtn < b:
        return -pre_rtn
    else:
        return 0

for x_col in back_df.columns[1:]:
    back_df[f'{x_col}_sta_rtn'] = back_df[x_col].apply(strategy)
```
